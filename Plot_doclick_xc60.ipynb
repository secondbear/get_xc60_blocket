```python
from re import sub
import re
#rec = re.compiler()'[^\d.]') 

#[sub(r'[^\d.]', '', x) for x in df_cars.price.to_list()]
df_cars.price = df_cars.price.replace(r'[^\d.]', '', regex=True)
```

```python
df_cars.text = 0
booleanDictionary = {True: 'TRUE', False: 'FALSE'}
df_cars["text"] = df_cars.make + "<br>"
df_cars["text"] = df_cars["text"]+ "drag:"
df_cars["text"] = df_cars["text"] +df_cars.drag.map(booleanDictionary)+" <br> "
```

```python
df_cars["text"] = (df_cars["text"] + "year:"+ df_cars.year.astype(str)+"<br>color:"+
                    df_cars.color)
```

```python
df_cars["text"] = df_cars["text"]+"<br>mil:"+df_cars.mil_low.astype(str) 
```

```python
#map color
color_map = {"gray":"grå", "black":"svart", "white":"vit", "silver":"silver", "blue":"blå",
             "red":"röd", "brown":"brun"}
df_cars["color_car"] = df_cars.color.apply(lambda x:  [ key  for key,val in color_map.items() if val in x.lower() ][0])
df_cars[["color", "color_car"]]
```

```python

fig = go.FigureWidget()

scatter = fig.add_trace(go.Scatter(x=df_cars["mil_low"], mode='markers',
                                   y=df_cars["price"],marker_color=df_cars["color_car"],
                                   text=df_cars["text"]))
hover_data =  ["drag", "kamera","navi", "gps", "taklucka"]
hover_data.extend(["year", "color"])

#fig = px.scatter(df_cars, x="mil_low", y="price",color="make", hover_data=hover_data)
#fig.show()
fig.update_traces(marker=dict(size=df_cars.year-2010,
                              line=dict(width=1,
                                        color='black')),
                  selector=dict(mode='markers'))


for model in df_cars.make.unique():
        # calculate polynomialmodel
    df_make = df_cars[df_cars.make ==model]
    #scipy.optimize.curve_fit
    z = np.polyfit(df_make.mil_low.sort_values(), df_make.price.sort_values(ascending=False), 1)
    f = np.poly1d(z)
    
    
    # calculate new x's and y's
    if model == "XC60":
        colorline = "blue"
    if model == "Outlander":
        colorline = "green"
    if model == "Optima":
        colorline = "red"
        
    x_new = np.linspace(df_make.mil_low.sort_values().iloc[0], df_make.mil_low.sort_values().iloc[-1], 50)
    y_new = f(x_new)
    trace2 = go.Scatter(
                      x=x_new,
                      y=y_new,
                      mode='lines',
                      marker= plotly.graph_objs.scatter.Marker(color=colorline),
                      name=model
                      
                      )
    fig.add_trace(trace2)
    
    
def do_click(trace, points, state):
    if points.point_inds:
        ind = points.point_inds[0]
        url = df_cars.ad_url.iloc[ind]
        webbrowser.open_new_tab(url)
#------------ test----------      
#------------------------
fig.layout.hovermode = 'closest'

fig.update_layout(
    title="Miltal mot pris, XC60 hela Sverige",
    xaxis_title="miltal angivet (mil)",
    yaxis_title="pris (kr)",
   
    font=dict(
        family="Courier New, monospace",
        size=14,
        color="Grey"
    )
)
scatter = fig.data[0]
scatter.on_click(do_click,True)
fig
```
